```python
def abcd(a,b,le):
    temp=0
    c_3='['+' '*le+']'
    c_2='('+' '*le+')'
    
    while True:
        if c_3 in a:
            i=a.index(c_3)
            for j in range(le):
                b[i+le-j-1]=b[i+le-j-1]+b[i+le-j]
                b[i+le-j]=0
            if le==0:
                b[i]+=3
            else:
                b[i]*=3
            a=a[:i]+' '*(le+2)+a[i+le+2:]
        elif c_2 in a:
            i=a.index(c_2)
            for j in range(le):
                b[i+le-j-1]=b[i+le-j-1]+b[i+le-j]
                b[i+le-j]=0
            if le==0:
                b[i]+=2
            else:
                b[i]*=2
            a=a[:i]+' '*(le+2)+a[i+le+2:]
        else:
            break
    if le<len(a)-3:        
        b=abcd(a,b,le+2) 
    return b

def check(a):
    stack=[]
    try:
        for i in a:
            if i=='(':
                stack+=[2]
            elif i=='[':
                stack+=[3]
            elif i==']':
                if stack[-1]==3:
                    del stack[-1]
                else :
                    return 0
            elif i==')':
                if stack[-1]==2:
                    del stack[-1]
                else :
                    return 0
            else :
                return 0
        if len(stack)==0:
            return True
    except:
        return 0

import sys

a=input()
b=[0 for i in range(len(a))]
if check(a)==True:
    print(sum(abcd(a,b,le=0)))
else :
    print(0)
```
